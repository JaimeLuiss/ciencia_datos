```python
# Mostrar información inicial
print("Información inicial del dataset:")
print(df.info())
print("\nDatos nulos por columna:")
print(df.isnull().sum())
```

```python
# Crear copias del dataset para diferentes estrategias
df_median = df.copy()  # Estrategia 1: Llenar con mediana/moda
df_mean = df.copy()    # Estrategia 2: Llenar con media/moda
df_drop = df.copy()    # Estrategia 3: Eliminar filas con NaN en 'Age' o 'Embarked'
```

```python
# --- Estrategia 1: Llenar con mediana (age) y moda (embarked), eliminar deck ---
df_median.drop('deck', axis=1, inplace=True)  # Cambiar 'Cabin' por 'deck'
age_median = df_median['age'].median()  # Cambiar 'Age' por 'age'
df_median['age'].fillna(age_median, inplace=True)
df_median['age'] = df_median['age'].astype(int)
embarked_mode = df_median['embarked'].mode()[0]  # Cambiar 'Embarked' por 'embarked'
df_median['embarked'].fillna(embarked_mode, inplace=True)
```

```python
# --- Estrategia 2: Llenar con media (age) y moda (embarked), eliminar deck ---
df_mean.drop('deck', axis=1, inplace=True)
age_mean = df_mean['age'].mean()
df_mean['age'].fillna(age_mean, inplace=True)
df_mean['age'] = df_mean['age'].astype(int)
df_mean['embarked'].fillna(embarked_mode, inplace=True)
```

```python
# --- Estrategia 3: Eliminar filas con NaN en 'age' o 'embarked', eliminar deck ---
df_drop.drop('deck', axis=1, inplace=True)
df_drop.dropna(subset=['age', 'embarked'], inplace=True)
df_drop['age'] = df_drop['age'].astype(int)

# Verificar que no hay datos nulos en las versiones tratadas
print("\nDatos nulos después de Estrategia 1 (mediana):")
print(df_median.isnull().sum())
print("\nDatos nulos después de Estrategia 2 (media):")
print(df_mean.isnull().sum())
print("\nDatos nulos después de Estrategia 3 (eliminar):")
print(df_drop.isnull().sum())

# --- Análisis descriptivo comparativo ---
print("\nResumen estadístico - Estrategia 1 (mediana):")
print(df_median.describe())
print("\nResumen estadístico - Estrategia 2 (media):")
print(df_mean.describe())
print("\nResumen estadístico - Estrategia 3 (eliminar):")
print(df_drop.describe())
```

```python
# --- Visualización comparativa de la distribución de 'age' ---
plt.figure(figsize=(12, 6))
sns.histplot(df_median['age'], bins=30, kde=True, label='Mediana', color='blue', alpha=0.5)
sns.histplot(df_mean['age'], bins=30, kde=True, label='Media', color='green', alpha=0.5)
sns.histplot(df_drop['age'], bins=30, kde=True, label='Eliminar NaN', color='red', alpha=0.5)
plt.title('Distribución de Edades según Estrategia de Tratamiento')
plt.legend()
plt.show()
```

```python
# --- Comparación de tasas de supervivencia para el perfil de Jack ---
jack_profile_median = df_median[(df_median['sex'] == 'male') &
                                (df_median['pclass'] == 3) &
                                (df_median['age'].between(15, 25))]
jack_profile_mean = df_mean[(df_mean['sex'] == 'male') &
                            (df_mean['pclass'] == 3) &
                            (df_mean['age'].between(15, 25))]
jack_profile_drop = df_drop[(df_drop['sex'] == 'male') &
                            (df_drop['pclass'] == 3) &
                            (df_drop['age'].between(15, 25))]

survival_rate_median = jack_profile_median['survived'].mean()
survival_rate_mean = jack_profile_mean['survived'].mean()
survival_rate_drop = jack_profile_drop['survived'].mean()

print(f"\nTasa de supervivencia para Jack (Estrategia 1 - Mediana): {survival_rate_median:.2%}")
print(f"Tasa de supervivencia para Jack (Estrategia 2 - Media): {survival_rate_mean:.2%}")
print(f"Tasa de supervivencia para Jack (Estrategia 3 - Eliminar): {survival_rate_drop:.2%}")
```

```python
# --- Visualización de supervivencia por estrategia ---
plt.figure(figsize=(10, 6))
sns.barplot(x=['Mediana', 'Media', 'Eliminar'],
            y=[survival_rate_median, survival_rate_mean, survival_rate_drop])
plt.title('Tasa de Supervivencia para el Perfil de Jack por Estrategia')
plt.ylabel('Tasa de Supervivencia')
plt.show()
```
