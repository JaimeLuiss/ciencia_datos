```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#1000 partidas
np.seed(42)
num_partidas = 100
fechas = pd.date_range("2025-01-01", periods=num_partidas, freq="D").strftime('%Y-%m-%d')
kills = np.poisson(12, num_partidas)  # Media de 12 kills por partida
muertes = np.poisson(10, num_partidas)  # Media de 10 muertes por partida
ranked = np.choice([True, False], size=num_partidas, p=[0.6, 0.4])  # 60% partidas ranked

# Crear un DataFrame con los datos
df = pd.DataFrame({
    "Fecha": fechas,
    "Kills": kills,
    "Muertes": muertes,
    "Ranked": ranked
})

# Filtrar solo las partidas clasificatorias (Ranked)
df_ranked = df[df["Ranked"] == True]

# Calcular el promedio de kills y muertes para las partidas clasificatorias
promedio_kills_ranked = df_ranked["Kills"].mean()
promedio_muertes_ranked = df_ranked["Muertes"].mean()

# Mostrar el promedio
print(f"Promedio de Kills por partida clasificatoria: {promedio_kills_ranked:.2f}")
print(f"Promedio de Muertes por partida clasificatoria: {promedio_muertes_ranked:.2f}")

plt.figure(figsize=(12, 6))

# Gráfico de barras de kills y muertes por partida clasificatoria
plt.bar(df_ranked["Fecha"], df_ranked["Kills"], color='green', label='Kills', alpha=0.7, width=0.4, align='center')
plt.bar(df_ranked["Fecha"], df_ranked["Muertes"], color='red', label='Muertes', alpha=0.7, width=0.4, align='edge')

# Añadir líneas de promedio
plt.axhline(y=promedio_kills_ranked, color='green', linestyle='--', label=f'Promedio de Kills: {promedio_kills_ranked:.2f}')
plt.axhline(y=promedio_muertes_ranked, color='red', linestyle='--', label=f'Promedio de Muertes: {promedio_muertes_ranked:.2f}')

# Añadir detalles al gráfico
plt.title('Kills vs Muertes por Partida (Clasificatorias)', fontsize=16)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos manuales del jugador
data = {
    'DPI': [800],  # Valor de DPI
    'Sensibilidad': [0.315],  # Sensibilidad
    'eDPI': [800 * 0.315],  # eDPI calculado como DPI * Sensibilidad
    'Minimapa Tamaño': [0.898],  # Tamaño del minimapa
    'Minimapa Zoom': [0.792],  # Zoom del minimapa
    'Framerate': [144],  # Suposición de un framerate común
    'Calidad Gráfica': [0],  # Calidad gráfica en formato numérico (0 para Low, 1 para Medium, 2 para High)
}

# Crear DataFrame
df = pd.DataFrame(data)

# Generar la matriz de correlación
corr_matrix = df.corr()

# Mostrar la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación de Configuraciones')
plt.show()
```
